## Предобработка данных

### Заполнение пропусков

Начнем с поиска пропущенных значений методом isna.

```python
df.isna().sum()
```

Пропущенных значений очень много. Начнем с is_apartment — апартаменты. 20924 пропущенных значений скорее всего, если не указано, значит это не апартаменты. Заполним пропущенные значения параметром False. Проверим есть ли пропущенные значения после замены.

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
df['is_apartment'].isna().sum()
```

balcony число балконов.  11519 пропущенных значений скорее всего, если не указано, значит ,балконов. Заполним пропущенные значения параметром False. Проверим есть ли пропущенные значения после замены.

```python
df['balcony'] = df['balcony'].fillna(0)
df['balcony'].isna().sum()
```

ceiling_height, living_area, kitchen_area и floors_total численные значения высоты потолков, площадей жилой и кухонной зон, а также общего число этажей. Разумным будет заменить пропущенные значения медианным значением. После замены проверим число пропусков в столбцах.

В столбце rooms есть значения с 0 квартир. Посмотрим, что это за объекты.

```python
df[df['rooms']<1]
```

Заменим значения в таких столбцах на медианное по столбцу rooms. В объеслениях указано много параметров, но по имеющейся информации мы не сможем восстановить скольки комнатная квратира в объявлении. После замены выведем минимальное значение по столбцу rooms в dataframe.

```python
df['rooms'] = df['rooms'].replace(0, df['rooms'].median())
df['rooms'].min()
```

```python
df_cloumns_pass=['ceiling_height', 'floors_total']
for column in df_cloumns_pass:
    df[column] = df[column].fillna(df[column].median())
df.isna().sum()
```

```python
df_columns_pass = ['kitchen_area', 'living_area']
for column in df_columns_pass:
    koef = df[column].sum()/df['total_area'].sum()
    df[column] = df[column].fillna(koef*df['total_area'])
df.isna().sum()    
```

Оставщиеся пропуски оставим. Заполнить их какими-то усредненными или нулевыми значениями не имеет смысла.Однако и удалять эти строки также неразумно. В некоторых столбцах, например в parks_nearest доля пропущенных значений очень высока. Мы потерям большую часть выборки.

## Расчёты и добавление результатов в таблицу

### Расчет цены кв метра

Цена за кв мемет получается путем деления стоимости квартиры (столбец last_price) на площадь квартиры(столбец total_area).
<br>Добавим столбец price_square_meter и добавим в него данные расчета. Выведем значения столбца.

```python
df['price_square_meter']=df['last_price']/df['total_area']
df['price_square_meter'].value_counts(ascending=False)
```

### Расчет дня недели, месяца и года публикации объявления

Определим день недели в столбце first_day_exposition методом dt.weekday. Создадим столбцы месяц публикации и год: month и year соответственно. Выведем df на экран.

```python
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
df
```

### Категоризируем этаж квартиры floor

Создадим функцию для этой цели.

```python
def floor_group(dataframe):
    total_floor = dataframe['floors_total']
    floor =  dataframe['floor']
    if floor == 1:
        return 'первый'
    elif floor == total_floor:
        return 'последний'
    elif 1 < floor < total_floor:
        return 'другой'
```

Применим функцию для категоризации квартир по этажам(первый, последний и другой).

```python
df['floor_category'] = df.apply(floor_group, axis = 1)
df['floor_category'].value_counts()
```

### Расчет соотношения жилой и общей площади, а также отношения площади кухни к общей.

Для расчета соотношения жилой и общей площади необходимо разделить значения living_area на total_area. 
<br>Значения сохраним в столбце ratio_living_total. 

```python
df['ratio_living_total']=df['living_area']/df['total_area']
df['ratio_living_total']=df['ratio_living_total']
df['ratio_living_total'].value_counts()
```

Для расчета отношения площади кухни к общей необходимо разделить значения kitchen_area  на total_area.
<br>Значения сохраним в столбце ratio_kitchen_total. 

```python
df['ratio_kitchen_total']=df['kitchen_area']/df['total_area']
df['ratio_kitchen_total'].value_counts()
```

Средняя цена квадратного метра это отношение средней стоимости квартиры в населенном пункте на среднюю площадь квартиры.

Переименуем названия столбцов, избавившись от двойных названий.

### Зависимость стоимости жилья от удаленности от центра в СПб

Выделим квартиры в Санкт-Петербурге. 

```python
df=df.query('locality_name == "Санкт-Петербург"')
```

Создадим столбец с расстоянием до центра в километрах. Оставим в dataframe только те строки, где есть значения до центра и округлим расстояния до центра до целых значений. Выведем значения столбца методом value_counts() для того, чтобы убедиться, что все значения целые.

```python
df['cityCenters_nearest_km'] = (df['cityCenters_nearest']/1000)
df=df.query('cityCenters_nearest_km>0')
df['cityCenters_nearest_km'] = df['cityCenters_nearest_km'].apply(np.ceil)
df['cityCenters_nearest_km'].value_counts()
```

Посчитаем среднюю цену для каждого километра. Для этого сгруппируем значения по удаленности и расситаем средние площади и стоимости. Путем деления последних вычислим среднюю стоимость метра и добавим ее в dataframe.

```python
df_cityCenters_nearest = df.groupby('cityCenters_nearest_km')['last_price','total_area'].agg(['count', 'mean'])
df_cityCenters_nearest.columns=[('count'),
            ('last_price_mean'),
            ('count_area'),
            ('total_area_mean')]
df_cityCenters_nearest = df_cityCenters_nearest[['count','last_price_mean','total_area_mean']]
df_cityCenters_nearest['meter_price_mean'] = df_cityCenters_nearest['last_price_mean'] / df_cityCenters_nearest['total_area_mean']  
df_cityCenters_nearest
```

Пострим график зависимости цены от удалённости от центра.

```python
df_cityCenters_nearest['meter_price_mean'].plot()
```

 По полученным данным график сильно меняется в близи центра в зоне до 10 км. Таким образом, опередлим центр как зону в 10 км.
