### Стандартизация данных

Проведем стандартизацию данных с пмощью StandardScaler, чтобы избежать проблем с масштабами признаков.

```python
numeric=['RowNumber', 'CustomerId', 'CreditScore', 'Age', 'Tenure', 'Balance',
       'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']
```

```python
scaler = StandardScaler()
scaler.fit(features_train,target_train)
features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric]  = scaler.transform(features_test[numeric])
```

### Логическая регрессия

Попробуем использовать модель логической регрессиии LogisticRegression. Для подбора параметров воспользуемся RandomizedSearchCV. Для удобства расчета f1 и roc_auc напишем функцию.

```python
def modeling(model,param,features_tr,target_tr,features_val,target_val):
    random_search = RandomizedSearchCV(model, param,
                                   n_iter=7, cv=5, scoring='f1')

    random_search.fit(features_tr,target_tr)
    predicted_valid = random_search.predict(features_val)
    f1_score_ = f1_score(target_valid, predicted_valid)
    
    probabilities_valid = random_search.predict_proba(features_valid)
    precision, recall, thresholds = precision_recall_curve(target_valid,
                                                       probabilities_valid[:, 1])
    probabilities_one_valid = probabilities_valid[:, 1]
    auc_roc_ = roc_auc_score(target_valid, probabilities_one_valid)
    return f1_score_,auc_roc_,random_search.best_params_, random_search.best_estimator_
```

```python
param_grid_LR = {
    'solver': ['lbfgs','liblinear'],
    'max_iter': range(100, 2000,100)}

LR=LogisticRegression(random_state=12345)

f1_score_LR, auc_roc_LR,best_params_LR, best_model_LR = modeling(LR,param_grid_LR,
                       features_train,target_train,features_valid,target_valid)

print(f'F1 логической регрессии {f1_score_LR}')
print(f'AUC-ROC логической регрессии {auc_roc_LR}')
print(f'Оптимальные параметры логической регрессии {best_params_LR}')
print(f'Модель логической регрессии {best_model_LR}')
```

### Модель решающего дерева

Попробуем использовать модель решающего дерева DecisionTreeClassifier. Для подбора параметров воспользуемся RandomizedSearchCV.

```python
param_grid_DT = {'max_depth': range(1,20),
             'min_samples_split' :range (0,12),
             'min_samples_leaf':range (0,12)}

DT = DecisionTreeClassifier(random_state=12345)

f1_score_DT,auc_roc_DT,best_params_DT, best_model_DT = modeling(DT,param_grid_DT,
                       features_train,target_train,features_valid,target_valid)

print(f'F1 решающего дерева {f1_score_DT}')
print(f'AUC-ROC решающего дерева {auc_roc_DT}')
print(f'Оптимальные параметры решающего дерева {best_params_DT}')
print(f'Модель решающего дерева {best_model_DT}')
```

### Модель случайного леса

Попробуем использовать модель решающего дерева RandomForestClassifier. Для подбора параметров воспользуемся RandomizedSearchCV.

```python
param_grid_RF = {'n_estimators': range (100,2000,100),
             'max_depth': range(1,20),
             'min_samples_split' :range (0,12),
             'min_samples_leaf':range (0,12)}

RF = RandomForestClassifier(random_state=12345)

f1_score_RF,auc_roc_RF,best_params_RF, best_model_RF = modeling(RF,param_grid_RF,
                       features_train,target_train,features_valid,target_valid)

print(f'F1 случайного леса {f1_score_RF}')
print(f'AUC-ROC случайного леса {auc_roc_RF}')
print(f'Оптимальные параметры решающего дерева {best_params_RF}')
print(f'Модель решающего дерева {best_model_RF}')
```

### Увеличение выборки.

Напишем функцию для увеличения объектов выборки. Разделим выборку на отризательные и положительные объекты. Положительные объекты скопируем и увеличим выборки.  

```python
def upsample(features,target,repeat):
    features_zeros = features[target==0]
    features_ones = features[target==1]
    target_zeros = target[target==0]
    target_ones = target[target==1]
    features_upsampled = pd.concat([features_zeros]+[features_ones]*repeat)
    target_upsampled = pd.concat([target_zeros]+[target_ones]*repeat)
    return features_upsampled,target_upsampled
```

Вызовом функции увеличения выборки, создаются новая обучающая выборка. Перемешаем объекты. 

```python
features_upsampled,target_upsampled = upsample(features_train, target_train, 4)
features_upsampled = shuffle(features_upsampled, random_state=12345)
target_upsampled = shuffle(target_upsampled, random_state=12345)
```

Проведем обучение моделей по новым данным для всех моделей.

### Уменьшение выборки

Напишем функцию для уменьшения объектов частого класса выборки. Разделим выборку на отризательные и положительные объекты. Случайным образом отбросим отрицательные и уменьшим итоговые выборки.

```python
def downsample(features,target,fraction):
    features_zeros = features[target==0]
    features_ones = features[target==1]
    target_zeros = target[target==0]
    target_ones = target[target==1]
    features_downsampled = pd.concat([features_zeros.sample(frac=fraction,random_state=12345)]+[features_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction,random_state=12345)]+[target_ones])
    return features_downsampled,target_downsampled
```

Вызовом функции увеличения выборки, создаются новая обучающая выборка. Перемешаем объекты. 

```python
features_downsampled,target_downsampled = downsample(features_train, target_train, 0.1)
features_downsampled = shuffle(features_downsampled, random_state=12345)
target_downsampled = shuffle(target_downsampled, random_state=12345)
```
