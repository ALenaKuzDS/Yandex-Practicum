### Проверка расчета эффективности обогащения

Нужно проверить, что эффективность обогащения рассчитана правильно.<br>Вычислим её на обучающей выборке для признака rougher.output.recovery. 
<br> Расчёт эффективности Recovery = C * (F - T)/ F * (C - T)
<br> C — доля золота в концентрате после флотации/очистки - rougher.output.concentrate_au
<br> F — доля золота в сырье/концентрате до флотации/очистки - rougher.input.feed_au
<br> T — доля золота в отвальных хвостах после флотации/очистки - rougher.output.tail_au

```python
C = df_train['rougher.output.concentrate_au'] 
F = df_train['rougher.input.feed_au'] 
T = df_train['rougher.output.tail_au'] 
df_train['rougher_recovery_cal'] =  ((C * (F - T))/ (F * (C - T))) * 100
print(df_train['rougher_recovery_cal'])
```

Найдем MAE между расчётами и значением признака rougher.output.recovery. 

```python
mean_absolute_error(df_train['rougher_recovery_cal'],df_train['rougher.output.recovery'])
```

**Вывод по rougher.output.recovery** признак рассчитан верно. Значение МАЕ 9.73512347450521e-15 - очень низкое.

## Модель

### Функция для вычисления итоговой sMAPE

Напишим функцию для вычисления итоговой sMAPE.

```python
def sMAPE_func(predicted, target):
    total__=0
    total__ = abs(target-predicted)/(2*(abs(target) + abs(predicted))).mean()
    return total__*100
```

```python
def TOTAL_sMAPE_function(predicted, target):
    predicted = pd. DataFrame(predicted)
    target = pd. DataFrame(target)
    rougher_recovery_sMAPE = sMAPE_func(predicted.iloc[:,0],target.iloc[:,0])
    final_recovery_sMAPE = sMAPE_func(predicted.iloc[:,1],target.iloc[:,1])
    return min(25*rougher_recovery_sMAPE+ 75*final_recovery_sMAPE)
```
